```python
from transformers import AutoModel, AutoTokenizer

# 1. KoBERT 토크나이저와 모델 불러오기
kobert_tokenizer = AutoTokenizer.from_pretrained("skt/kobert-base-v1", use_fast=False)
kobert = AutoModel.from_pretrained("skt/kobert-base-v1")

# 2. 추가할 새로운 토큰 정의
new_tokens = ['새로운단어1', '새로운단어2']

# 3. 기존 토크나이저에 없는 토큰만 추가 (중복 방지)
existing_tokens = set(kobert_tokenizer.get_vocab().keys())
tokens_to_add = [token for token in new_tokens if token not in existing_tokens]

# 4. 새로운 토큰 추가
if tokens_to_add:
    kobert_tokenizer.add_tokens(tokens_to_add)
    # print(f"새로운 토큰 {}개가 추가되었습니다: "{tokens_to_add}")

    # 모델의 임베딩 레이어 크기 확장 (중요)
    kobert.resize_token_embeddings(len(kobert_tokenizer))

# 5. 추가된 토큰으로 테스트
text = "안녕하세요, 새로운단어1를 추가한 KoBERT 모델입니다."

encoded_input = kobert_tokenizer.tokenize(text, return_tensors='pt')

# 6. 인코딩된 input의 최대 인덱스가 모델의 임베딩 크기를 초과하지 않는지 확인
vocab_size = kobert.config.vocab_size
print(f"모델의 vocab 크기: {vocab_size}")
print(f"입력된 텐서 최대 인덱스: {encoded_input['input_ids'].max()}")

# 7. 모델 출력 확인
output = kobert(**encoded_input)
print(output)
```
